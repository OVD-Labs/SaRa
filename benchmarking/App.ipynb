```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import saraRC1 as sara
```

```python
im = cv2.imread('./images/3_colour.jpeg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
sm_itti = sara.return_saliency(im, generator='itti')
sm_dg = sara.return_saliency(im, generator='deepgaze')
```

```python
# Show original
plt.figure()
plt.imshow(im)
plt.xticks([])
plt.yticks([])

plt.tight_layout()
plt.show()


# Show saliency maps
plt.figure()
plt.subplot(121)
plt.imshow(sm_itti, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(sm_dg, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.tight_layout()
plt.show()


# Show saliency maps overlayed on original
plt.figure()
plt.subplot(121)
plt.imshow(im)
plt.imshow(sm_itti, cmap='jet', alpha=0.5)
plt.title('Itti')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(im)
plt.imshow(sm_dg, cmap='jet', alpha=0.5)
plt.title('DeepGaze IIE')
plt.xticks([])
plt.yticks([])
plt.tight_layout()
plt.show()
```
